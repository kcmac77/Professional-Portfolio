```python
url= 'https://www.the-numbers.com/movies/franchise/Marvel-Cinematic-Universe#tab=summary'
```

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

response = requests.get(url, headers=headers)

print(response.text) 
```

```python
soup = BeautifulSoup(response.text,"html.parser")
tables = soup.find('table') # in
data = []
```

```python
for idx, table in enumerate(tables):
    print(f"Table {idx+1}")
    print(table)
    print("\n\n")
```

```python
for row in tables.find_all('tr'):
    # Initialize an empty dictionary to store row data
    row_data = {}
    
    # Get all columns ('td') in the current row
    columns = row.find_all('td')
    
    # Check if the row contains the expected number of columns
    if len(columns) >= 2:  # Ensure there are at least two columns in the row
        # Extract data from each column and store it in the dictionary
        row_data['Release Date'] = columns[0].get_text().strip()
        
        # Find the <a> tag inside the second column (for Movie Title)
        movie_title_elem = columns[1].find('a')
        if movie_title_elem:
            row_data['Movie Title'] = movie_title_elem.get_text().strip()
        else:
            row_data['Movie Title'] = 'N/A'  # Set a default value if <a> tag not found
        
        # Optional: Extract additional data columns if needed
        if len(columns) >= 6:
            row_data['Data1'] = columns[2].get_text().strip()
            row_data['Data2'] = columns[3].get_text().strip()
            row_data['Data3'] = columns[4].get_text().strip()
            row_data['Data4'] = columns[5].get_text().strip()
        else:
            # Set default values or handle missing columns
            row_data['Data1'] = 'N/A'
            row_data['Data2'] = 'N/A'
            row_data['Data3'] = 'N/A'
            row_data['Data4'] = 'N/A'
        
        # Append the row data to the list
        data.append(row_data)

# Display the extracted data
for idx, entry in enumerate(data):
    print(f"Row {idx + 1}:")
    for key, value in entry.items():
        print(f"{key}: {value}")
    print()  # Print a blank line between rows
```

```python
import csv
csv_filename = 'marvel_movies.csv'

# Write the extracted data to a CSV file
with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    # Write header row
    writer.writerow(['Release Date', 'Movie Title', 'Data1', 'Data2', 'Data3', 'Data4'])
    # Write data rows
    writer.writerows(row_data)

print(f"CSV file '{csv_filename}' has been created successfully.")
```

```python
csv_file = 'output.csv'

# List of keys to be used as column headers
fieldnames = ['Release Date', 'Movie Title', 'Data1','Data2','Data3','Data4']

# Open the CSV file in write mode
with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    # Create a DictWriter object
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write the header row based on the fieldnames
    writer.writeheader()
    
    # Write data rows based on the dictionary
    for row in data:
        writer.writerow(row)

print(f"CSV file '{csv_file}' has been created successfully.")
```

```python
def remove_duplicate_rows(input_file, output_file):
    unique_rows = set()  # Set to store unique rows

    # Read CSV file and identify unique rows
    with open(input_file, 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        header = next(reader)  # Read header row
        unique_rows.add(tuple(header))  # Add header as tuple (considered unique)

        for row in reader:
            row_tuple = tuple(row)
            unique_rows.add(row_tuple)  # Add row as tuple

    # Write unique rows to a new CSV file
    with open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(header)  # Write header row

        for row in unique_rows:
            writer.writerow(row)

    print(f"Duplicate rows removed. Unique rows saved to '{output_file}'.")

# Example usage:
input_csv = 'output.csv'
output_csv = 'output_unique.csv'

# Call the function to remove duplicate rows
remove_duplicate_rows(input_csv, output_csv)
```
