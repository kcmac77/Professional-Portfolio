```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the Wikipedia page containing the table
url = 'https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films'

# Send a GET request to fetch the HTML content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text
else:
    print(f"Failed to retrieve the page: {url}")
    exit()

# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table element with a specific class
table = soup.find('table', class_='wikitable')  # Adjust class or ID as needed

# Check if the table was found
if table:
    # Specify the CSV file path
    csv_file_path = 'marvel_movies.csv'

    # Open the CSV file in write mode
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
        # Create a CSV writer object
        csv_writer = csv.writer(csvfile)

        # Iterate over rows in the table
        for row in table.find_all('tr'):
            # Extract data from each cell in the row
            cells = row.find_all(['th', 'td'])
            if cells:
                # Extract the text content of each cell and write to CSV
                row_data = [cell.text.strip() for cell in cells]
                csv_writer.writerow(row_data)

    print(f"CSV file '{csv_file_path}' has been successfully created.")
else:
    print("Table not found on the page.")
```

```python
html_contents= '''</tbody></table>
<h3><span class="mw-headline" id="The_Multiverse_Saga">The Multiverse Saga</span></h3>
<link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1033289096"><div role="note" class="hatnote navigation-not-searchable">Main articles: <a href="/wiki/Marvel_Cinematic_Universe:_Phase_Four" title="Marvel Cinematic Universe: Phase Four">Marvel Cinematic Universe: Phase Four</a>, <a href="/wiki/Marvel_Cinematic_Universe:_Phase_Five" title="Marvel Cinematic Universe: Phase Five">Phase Five</a>, and <a href="/wiki/Marvel_Cinematic_Universe:_Phase_Six" title="Marvel Cinematic Universe: Phase Six">Phase Six</a></div>
<p>The films from Phase Four through Phase Six are collectively known as "The Multiverse Saga".<sup id="cite_ref-SDCC2022_22-1" class="reference"><a href="#cite_note-SDCC2022-22">&#91;21&#93;</a></sup><sup id="cite_ref-PhaseSixIGN_69-0" class="reference"><a href="#cite_note-PhaseSixIGN-69">&#91;68&#93;</a></sup> The Phases also include <a href="/wiki/List_of_Marvel_Cinematic_Universe_television_series#Marvel_Studios" title="List of Marvel Cinematic Universe television series">multiple series</a> and <a href="/wiki/Marvel_Studios_Special_Presentations" title="Marvel Studios Special Presentations">two specials</a> streaming on <a href="/wiki/Disney%2B" title="Disney+">Disney+</a>.<sup id="cite_ref-DigitalSpyPhases_25-4" class="reference"><a href="#cite_note-DigitalSpyPhases-25">&#91;24&#93;</a></sup>
<link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1211901571">
</p>
<table class="wikitable plainrowheaders defaultcenter col2left" style="width: 99%;">
<caption><link rel="mw-deduplicated-inline-style" href="mw-data:TemplateStyles:r1152813436"><span class="sr-only">Films in the Multiverse Saga of the Marvel Cinematic Universe</span>
</caption>
<tbody><tr>
<th scope="col">Film
</th>
<th scope="col">U.S. release date
</th>
<th scope="col">Director(s)
</th>
<th scope="col">Screenwriter(s)
</th>
<th scope="col">Producer(s)
</th>
<th scope="col">Status
</th></tr>
<tr>
<th colspan="6" style="background-color: #ccccff;"><a href="/wiki/Marvel_Cinematic_Universe:_Phase_Four#Films" title="Marvel Cinematic Universe: Phase Four">Phase Four</a><span style="color:black;background-color:white;padding:1px;display:inline-block;line-height:50%"><sup id="cite_ref-DigitalSpyPhases_25-5" class="reference"><a href="#cite_note-DigitalSpyPhases-25">&#91;24&#93;</a></sup></span>
</th></tr>


<tr>
<th scope="row"><i><a href="/wiki/Black_Widow_(2021_film)" title="Black Widow (2021 film)">Black Widow</a></i>
</th>
<td>July&#160;9,&#160;2021<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2021-07-09</span>)</span><sup id="cite_ref-71" class="reference"><a href="#cite_note-71">&#91;b&#93;</a></sup>
</td>
<td><a href="/wiki/Cate_Shortland" title="Cate Shortland">Cate Shortland</a><sup id="cite_ref-BWShortland_72-0" class="reference"><a href="#cite_note-BWShortland-72">&#91;70&#93;</a></sup>
</td>
<td><a href="/wiki/Eric_Pearson" title="Eric Pearson">Eric Pearson</a><sup id="cite_ref-BWWriters_73-0" class="reference"><a href="#cite_note-BWWriters-73">&#91;71&#93;</a></sup>
</td>
<td><a href="/wiki/Kevin_Feige" title="Kevin Feige">Kevin Feige</a>
</td>
<td rowspan="7">Released
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Shang-Chi_and_the_Legend_of_the_Ten_Rings" title="Shang-Chi and the Legend of the Ten Rings">Shang-Chi and the Legend of the Ten Rings</a></i>
</th>
<td>September&#160;3,&#160;2021<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2021-09-03</span>)</span>
</td>
<td><a href="/wiki/Destin_Daniel_Cretton" title="Destin Daniel Cretton">Destin Daniel Cretton</a><sup id="cite_ref-DestinDanielCretton_74-0" class="reference"><a href="#cite_note-DestinDanielCretton-74">&#91;72&#93;</a></sup>
</td>
<td><a href="/wiki/Dave_Callaham" class="mw-redirect" title="Dave Callaham">Dave Callaham</a> &amp; Destin Daniel Cretton &amp; <a href="/wiki/Andrew_Lanham" title="Andrew Lanham">Andrew Lanham</a><sup id="cite_ref-ShangChiWriters_75-0" class="reference"><a href="#cite_note-ShangChiWriters-75">&#91;73&#93;</a></sup>
</td>
<td>Kevin Feige and<br /> <span class="nowrap"><a href="/wiki/Jonathan_Schwartz_(film_producer)" class="mw-redirect" title="Jonathan Schwartz (film producer)">Jonathan Schwartz</a></span>
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Eternals_(film)" title="Eternals (film)">Eternals</a></i>
</th>
<td>November&#160;5,&#160;2021<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2021-11-05</span>)</span>
</td>
<td><a href="/wiki/Chlo%C3%A9_Zhao" title="Chloé Zhao">Chloé Zhao</a><sup id="cite_ref-EternalsZhao_76-0" class="reference"><a href="#cite_note-EternalsZhao-76">&#91;74&#93;</a></sup>
</td>
<td>Chloé Zhao and Chloé Zhao &amp; Patrick Burleigh<br />and Ryan Firpo &amp; <a href="/wiki/Kaz_Firpo" title="Kaz Firpo">Kaz Firpo</a><sup id="cite_ref-EternalsWritingCredits_77-0" class="reference"><a href="#cite_note-EternalsWritingCredits-77">&#91;75&#93;</a></sup><sup id="cite_ref-EternalsWGAWritingCredits_78-0" class="reference"><a href="#cite_note-EternalsWGAWritingCredits-78">&#91;76&#93;</a></sup><sup id="cite_ref-79" class="reference"><a href="#cite_note-79">&#91;c&#93;</a></sup>
</td>
<td>Kevin Feige<br /> and <a href="/wiki/Nate_Moore_(film_producer)" class="mw-redirect" title="Nate Moore (film producer)">Nate Moore</a>
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Spider-Man:_No_Way_Home" title="Spider-Man: No Way Home">Spider-Man: No Way Home</a></i>
</th>
<td>December&#160;17,&#160;2021<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2021-12-17</span>)</span>
</td>
<td><a href="/wiki/Jon_Watts" title="Jon Watts">Jon Watts</a><sup id="cite_ref-WattsSM3Confirm_80-0" class="reference"><a href="#cite_note-WattsSM3Confirm-80">&#91;77&#93;</a></sup>
</td>
<td><a href="/wiki/Chris_McKenna_(writer)" title="Chris McKenna (writer)">Chris McKenna</a> &amp; <a href="/wiki/Erik_Sommers" title="Erik Sommers">Erik Sommers</a><sup id="cite_ref-FFHSequelWriters_81-0" class="reference"><a href="#cite_note-FFHSequelWriters-81">&#91;78&#93;</a></sup>
</td>
<td>Kevin Feige<br /> and <a href="/wiki/Amy_Pascal" title="Amy Pascal">Amy Pascal</a>
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Doctor_Strange_in_the_Multiverse_of_Madness" title="Doctor Strange in the Multiverse of Madness">Doctor Strange in the Multiverse of Madness</a></i>
</th>
<td>May&#160;6,&#160;2022<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2022-05-06</span>)</span>
</td>
<td><a href="/wiki/Sam_Raimi" title="Sam Raimi">Sam Raimi</a><sup id="cite_ref-RaimiDS2Confirm_82-0" class="reference"><a href="#cite_note-RaimiDS2Confirm-82">&#91;79&#93;</a></sup>
</td>
<td><a href="/wiki/Michael_Waldron" title="Michael Waldron">Michael Waldron</a><sup id="cite_ref-WaldronDS2_83-0" class="reference"><a href="#cite_note-WaldronDS2-83">&#91;80&#93;</a></sup>
</td>
<td>Kevin Feige
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Thor:_Love_and_Thunder" title="Thor: Love and Thunder">Thor: Love and Thunder</a></i>
</th>
<td>July&#160;8,&#160;2022<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2022-07-08</span>)</span>
</td>
<td><a href="/wiki/Taika_Waititi" title="Taika Waititi">Taika Waititi</a><sup id="cite_ref-WaititiThor4_84-0" class="reference"><a href="#cite_note-WaititiThor4-84">&#91;81&#93;</a></sup>
</td>
<td>Taika Waititi &amp; <a href="/wiki/Jennifer_Kaytin_Robinson" title="Jennifer Kaytin Robinson">Jennifer Kaytin Robinson</a><sup id="cite_ref-ThorLaTWritersCredits_85-0" class="reference"><a href="#cite_note-ThorLaTWritersCredits-85">&#91;82&#93;</a></sup>
</td>
<td>Kevin Feige and<br /> <a href="/wiki/Brad_Winderbaum" class="mw-redirect" title="Brad Winderbaum">Brad Winderbaum</a>
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Black_Panther:_Wakanda_Forever" title="Black Panther: Wakanda Forever">Black Panther: Wakanda Forever</a></i>
</th>
<td>November&#160;11,&#160;2022<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2022-11-11</span>)</span>
</td>
<td><a href="/wiki/Ryan_Coogler" title="Ryan Coogler">Ryan Coogler</a><sup id="cite_ref-CooglerBPSequel_86-0" class="reference"><a href="#cite_note-CooglerBPSequel-86">&#91;83&#93;</a></sup>
</td>
<td>Ryan Coogler &amp; <a href="/wiki/Joe_Robert_Cole" title="Joe Robert Cole">Joe Robert Cole</a><sup id="cite_ref-CooglerBPSequel_86-1" class="reference"><a href="#cite_note-CooglerBPSequel-86">&#91;83&#93;</a></sup><sup id="cite_ref-FeigeRTJul2021_87-0" class="reference"><a href="#cite_note-FeigeRTJul2021-87">&#91;84&#93;</a></sup>
</td>
<td>Kevin Feige<br /> and Nate Moore
</td></tr>

<tr>
<th scope="row"><i><a href="/wiki/Ant-Man_and_the_Wasp:_Quantumania" title="Ant-Man and the Wasp: Quantumania">Ant-Man and the Wasp: Quantumania</a></i>
</th>
<td>February&#160;17,&#160;2023<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2023-02-17</span>)</span>
</td>
<td><a href="/wiki/Peyton_Reed" title="Peyton Reed">Peyton Reed</a><sup id="cite_ref-AntMan3Confirmed_88-0" class="reference"><a href="#cite_note-AntMan3Confirmed-88">&#91;85&#93;</a></sup>
</td>
<td><a href="/wiki/Jeff_Loveness" title="Jeff Loveness">Jeff Loveness</a><sup id="cite_ref-AntMan3Loveness_89-0" class="reference"><a href="#cite_note-AntMan3Loveness-89">&#91;86&#93;</a></sup>
</td>
<td><a href="/wiki/Kevin_Feige" title="Kevin Feige">Kevin Feige</a> and<br /> <a href="/wiki/Stephen_Broussard" class="mw-redirect" title="Stephen Broussard">Stephen Broussard</a>
</td>
<td rowspan="3">Released
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/Guardians_of_the_Galaxy_Vol._3" title="Guardians of the Galaxy Vol. 3">Guardians of the Galaxy Vol. 3</a></i>
</th>
<td>May&#160;5,&#160;2023<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2023-05-05</span>)</span>
</td>
<td colspan="2"><a href="/wiki/James_Gunn" title="James Gunn">James Gunn</a><sup id="cite_ref-GunnVol3Rehired_90-0" class="reference"><a href="#cite_note-GunnVol3Rehired-90">&#91;87&#93;</a></sup>
</td>
<td rowspan="2">Kevin Feige
</td></tr>
<tr>
<th scope="row"><i><a href="/wiki/The_Marvels" title="The Marvels">The Marvels</a></i>
</th>
<td>November&#160;10,&#160;2023<span style="display:none">&#160;(<span class="bday dtstart published updated itvstart">2023-11-10</span>)</span>
</td>
<td><a href="/wiki/Nia_DaCosta" title="Nia DaCosta">Nia DaCosta</a><sup id="cite_ref-CMSequelDaCosta_91-0" class="reference"><a href="#cite_note-CMSequelDaCosta-91">&#91;88&#93;</a></sup>
</td>
<td>Nia DaCosta and <a href="/wiki/Megan_McDonnell" title="Megan McDonnell">Megan McDonnell</a> and Elissa Karasik<sup id="cite_ref-TheMarvelsWriters_92-0" class="reference"><a href="#cite_note-TheMarvelsWriters-92">&#91;89&#93;</a></sup>
</td></tr>'''
soup = BeautifulSoup(html_contents, 'html.parser')

# Find the table containing the movie data
table = soup.find('table', class_='wikitable')


# Initialize a list to store movie data dictionaries
movies_data = []

# Extract rows from the table
rows = table.find_all('tr')

# Process each row to extract movie details
for row in rows[2:]:  # Skip the first two rows (header rows)
    cells = row.find_all(['th', 'td'])
    if len(cells) >= 5:  # Ensure the row has the required number of cells
        movie_title = cells[0].text.strip()
        release_date = cells[1].text.strip()
        director = cells[2].text.strip()
        writers = cells[3].text.strip()
        producers = cells[4].text.strip()
        
        # Store movie details in a dictionary
        movie_details = {
            'Title': movie_title,
            'Release Date': release_date,
            'Director': director,
            'Writers': writers,
            'Producers': producers
        }
        
        # Append movie details dictionary to the list
        movies_data.append(movie_details)

# Define the path for the CSV file
csv_file_path = 'multiverse_films.csv'

# Define the fieldnames for the CSV file (column headers)
fieldnames = ['Title', 'Release Date', 'Director', 'Writers', 'Producers']

# Write the extracted movie data to a CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write the header row
    writer.writeheader()
    
    # Write each movie data as a row in the CSV file
    for movie_data in movies_data:
        writer.writerow(movie_data)

print(f"CSV file '{csv_file_path}' has been successfully created with {len(movies_data)} movies.")
```
